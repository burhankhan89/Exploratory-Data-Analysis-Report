```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('customer_shopping_data.csv')
```

```python
data['gender'].value_counts(normalize = True)
```

```python
counts = [1, 2, 3, 4, 5]
num_colors = len(counts)
colors = np.random.rand(num_colors, 3)
```

```python
#plt.figure(figsize = (20, 6))

ax = data["gender"].value_counts().plot(kind = 'bar', color = colors, rot = 0)
ax.set_xticklabels(('Female', 'Male'))

for p in ax.patches:
    ax.annotate(int(p.get_height()), (p.get_x() + 0.25, p.get_height() + 1), ha = 'center', va = 'bottom', color = 'black')
    ax.tick_params(axis = 'both', labelsize = 15)

plt.xlabel('Gender', weight = "bold", color = "#000000", fontsize = 14)
plt.ylabel('Counts', weight = "bold", color = "#000000", fontsize = 14)
plt.title('Gender Distribution', weight='bold', color='#000000', fontsize=14)
```

```python
unique_values = data['shopping_mall'].unique()
```

```python
print(unique_values)
```

```python
#plt.figure(figsize = (20, 6))
counts = data["shopping_mall"].value_counts()
explode = tuple([0.1 if i == 1 else 0 for i in range(len(counts))])
plt.figure(figsize=(8, 8))
counts.plot(kind='pie', fontsize=12, colors=colors, explode=explode, autopct='%.1f%%')
plt.title('Shopping Malls Foot Fall Categorizations', weight='bold', color='#000000', fontsize=14)
plt.xlabel('shopping_mall', weight="bold", color="#000000", fontsize=14, labelpad=20)
plt.ylabel('Counts', weight="bold", color="#000000", fontsize=14, labelpad=20)
plt.legend(labels=counts.index, loc="center left", bbox_to_anchor=(1, 0.5))  # Place legend to the right
plt.show()
```

```python
unique_values = data['category'].unique()
print(unique_values)
```

```python
category_counts = data["category"].value_counts()

plt.figure(figsize=(10, 6))
plt.hist(category_counts.index, weights=category_counts.values, bins=len(category_counts), align='mid', orientation='horizontal', color='skyblue', rwidth=0.7, edgecolor='black')

# Adding labels and title
plt.title('Classification of Sold Products', weight='bold', color='#000000', fontsize=14)
plt.xlabel('Frequency')
plt.ylabel('Category')

plt.yticks(range(len(category_counts)), category_counts.index, rotation=45)  # Set y-axis ticks and labels

# Adjust layout
plt.tight_layout()

# Showing the plot
plt.show()
```
